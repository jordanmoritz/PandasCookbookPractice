```python
def download_format_weather(year, month):
    url_template = 'http://climate.weather.gc.ca/climate_data/bulk_data_e.html?format=csv&stationID=10761&Year={year}&Month={month}&Day=1&timeframe=1&submit=Download+Data'
    url = url_template.format(year=year, month=month)
    
    weather_data = pd.read_csv(url, skiprows=15, index_col='Date/Time', parse_dates=True, encoding='latin1')
    
    weather_data.columns = [col.replace('\xc2\xb0', '') for col in weather_data.columns]
    
#     weather_data.columns = [u'Year', u'Month', u'Day', u'Time', u'Temp C', u'Temp Flag', u'Dew Point C',
#                            u'Dew Point Flag', u'Rel Hum', u'Rel Hum Flag', u'Wind Dir', u'Wind Dir Flag',
#                            u'Wind Speed km/h', u'Wind Speed Flag', u'Visibility', u'Visibility Flag', u'Stn Press kPA',
#                            u'Stn Press Flag', u'Hmdx', u'Hmdx Flag', u'Wind Chill', u'Wind Chill Flag', u'Weather']
    
    weather_data = weather_data.dropna(axis=1, how='any')
    weather_data = weather_data.drop(['Year', 'Month', 'Day', 'Time'], axis=1)
    
    weather_data[:2]
    
    return weather_data
```

```python
weather_2015 = download_format_weather('2016', '3')

weather_2015[:5]
```

```python
data_by_month = [download_format_weather(y, m) for y in range (2012, 2014) for m in range(1, 13)]
```

```python
weather_two_years = pd.concat(data_by_month)
```
