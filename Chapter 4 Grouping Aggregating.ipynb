```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.fantasy'] = 'Comic-Sans MS'
plt.rcParams['figure.figsize'] = (15, 5)

pd.set_option('display.width', 5000)
pd.set_option('display.max_columns', 60)

file = 'pandas-cookbook-master/data/bikes.csv'
```

```python
bikes_df = pd.read_csv(file, sep=';', encoding='latin1', parse_dates=['Date'], dayfirst=True, index_col='Date')

berri_bikes_df = bikes_df[['Berri 1']].copy()

berri_bikes_df.plot()

berri_bikes_df[:5]
```

```python
berri_bikes_df.loc[:, 'Weekday'] = berri_bikes_df.index.weekday

berri_bikes_df[:5]
```

```python
weekday_berri_counts = berri_bikes_df.groupby(by='Weekday').aggregate(sum)

weekday_berri_counts
```

```python
weekday_berri_counts.index = ['Mon', 'Tues', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']

weekday_berri_counts.plot(kind='bar')
```

```python
bike_df = pd.read_csv(file, sep=';', encoding='latin1', parse_dates=['Date'], dayfirst=True, index_col='Date')
berri_bike_df = bike_df[['Berri 1']].copy()
berri_bike_df.loc[:, 'Month'] = berri_bike_df.index.month
month_berri_count = berri_bike_df.groupby(by='Month').aggregate(sum)
month_berri_count.index = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov']
month_berri_count.plot(kind='bar')
```
